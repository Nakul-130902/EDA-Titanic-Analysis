```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\ROYAL ENFIELD\OneDrive\Documents\Desktop\titanic (1).csv")
```

```python
df.head()
```

```python
print(df.info())
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python

median_age = df['age'].median()
df['age'].fillna(median_age, inplace=True)


mode_embarked = df['embarked'].mode()[0]
df['embarked'].fillna(mode_embarked, inplace=True)




print(df.info())
```

```python

 df.drop('deck', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python

mode_embark_town = df['embark_town'].mode()[0]
df['embark_town'].fillna(mode_embark_town, inplace=True)
df.isnull().sum()


```

# Data Cleaned!!!

```python
# Survived passengers count
sns.countplot(x='survived', data=df)
plt.title('Survival Count (0=Died, 1=Survived)')
plt.show()

# Gender count
sns.countplot(x='sex', data=df)
plt.title('Gender Count')
plt.show()

# Pclass count
sns.countplot(x='pclass', data=df)
plt.title('Passenger Class Count')
plt.figure(figsize=(10, 2)) 
plt.show()
```

```python
# Age ka distribution
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Fare ka distribution
sns.histplot(df['fare'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

Survival Vs Pclass

```python
sns.countplot(x='survived', hue='pclass', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

Survival VS Age

```python
sns.violinplot(x='pclass', y='age', hue='survived', data=df, split=True)
plt.title('Survival by Age and Class')
plt.show()
```

Correlation Matrix

```python
# Numerical columns ke beech correlation
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Survival Vs Gender

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

Survival vs. Passenger Class (pclass)

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
